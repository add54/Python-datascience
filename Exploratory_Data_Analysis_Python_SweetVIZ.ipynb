1. Read in the Penguins Dataset

```
import pandas as pd
penguins = pd.read_csv('http://raw.githubusercontent.com/dataprofessor/data/master/penguins_cleaned.csv')
```

```
penguins.head(2)
```

```
#Separating X and y
X = penguins.drop('species', axis=1)
y = penguins['species']
X
```

2. Analyze the Penguins DataSet

```
penguins
```

```
import sweetviz as sv
analyze_report = sv.analyze(penguins)
analyze_report.show_html('analyze.html', open_browser=False)
```

```
import IPython
IPython.display.HTML('analyze.html')
```

3. Side-by-Side Comparison of Train vs Test Set
.1 Data Split using 80/20 SplitRatio

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train
```

```
X_test
```

-Comparison Report

```
import sweetviz as sv
compare_report = sv.compare([X_train, 'train'], [X_test, 'test'])
compare_report.show_html('compare.html', open_browser=False)
```

```
import IPython
IPython.display.HTML('compare.html')
```
